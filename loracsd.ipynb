```python
import os

# --- ƒê·ªãnh nghƒ©a c√°c ƒë∆∞·ªùng d·∫´n ---
IMAGE_DIR = "/kaggle/working/SDXL_train"
METADATA_FILE = os.path.join(IMAGE_DIR, "metadata.jsonl")
OUTPUT_DIR = "/kaggle/working/SDXL_LoRA_model"

# ƒê∆∞·ªùng d·∫´n ƒë·∫øn c√°c model ƒë√£ ƒë√≠nh k√®m t·ª´ Kaggle Datasets
BASE_MODEL_PATH = "/kaggle/input/stabilityai-stable-diffusion-xl-base-1-0/sd_xl_base_1.0"
VAE_PATH = "/kaggle/input/madebyollin-sdxl-vae-fp16-fix/sdxl_vae.safetensors"

print("C√°c ƒë∆∞·ªùng d·∫´n ƒë√£ ƒë∆∞·ª£c thi·∫øt l·∫≠p.")
print(f"Th∆∞ m·ª•c ·∫£nh: {IMAGE_DIR}")
```

```python
import torch
import math
import json
import random
from transformers import CLIPTextModel, AutoTokenizer
from tqdm.notebook import tqdm
import numpy as np
import gc

print("--- B·∫Øt ƒë·∫ßu B∆∞·ªõc CSD: T√≠nh to√°n v√† L·∫•y m·∫´u l·∫°i Dataset ---")

# --- 1. N·∫°p c√°c model c·∫ßn thi·∫øt cho CSD ---
print("ƒêang t·∫£i Text Encoder c·ªßa SDXL ƒë·ªÉ t√≠nh CSD...")
device = "cuda" if torch.cuda.is_available() else "cpu"
text_encoder = CLIPTextModel.from_pretrained(
    "openai/clip-vit-large-patch14", torch_dtype=torch.float16
).to(device)
tokenizer = AutoTokenizer.from_pretrained(
    "openai/clip-vit-large-patch14"
)

# --- 2. Code CSD ---
def semantic_richness(text: str) -> float:
    inputs = tokenizer(text, return_tensors="pt", padding="max_length", truncation=True, max_length=77).to(device)
    with torch.no_grad():
        outputs = text_encoder(**inputs)
    attn = inputs['attention_mask'][0]
    idxs = torch.arange(attn.size(0), device=device)
    real = (attn == 1) & (idxs > 0) & (idxs < attn.sum() - 1)
    emb = outputs.last_hidden_state[0][real]
    if emb.shape[0] < 2:
        return 0.0
    return torch.var(emb, dim=0).sum().item()

def csd_score(text: str) -> float:
    n = len(text.split())
    if n == 0: return 0.0
    return float(semantic_richness(text) / math.log(1 + n))

# --- 3. ƒê·ªçc metadata, t√≠nh to√°n v√† chu·∫©n h√≥a CSD ---
# ƒê∆∞·ªùng d·∫´n file metadata trong m√¥i tr∆∞·ªùng Colab
METADATA_FILE = "/kaggle/working/SDXL_train/metadata.jsonl"
dataset = []
with open(METADATA_FILE, 'r') as f:
    for line in f:
        dataset.append(json.loads(line))

print(f"B·∫Øt ƒë·∫ßu t√≠nh CSD cho {len(dataset)} captions...")
for item in tqdm(dataset, desc="Calculating CSD Scores"):
    item['csd'] = csd_score(item['prompt'])

weights = [item['csd'] for item in dataset]
min_csd, max_csd = min(weights), max(weights)

final_weights = []
for w in weights:
    if (max_csd - min_csd) > 0:
        normalized_w = 0.1 + 0.9 * (w - min_csd) / (max_csd - min_csd)
    else:
        normalized_w = 1.0
    final_weights.append(normalized_w)
    
print("T√≠nh to√°n CSD v√† tr·ªçng s·ªë ho√†n t·∫•t.")

# --- 4. L·∫•y m·∫´u l·∫°i (Resample) dataset ---
print("B·∫Øt ƒë·∫ßu l·∫•y m·∫´u l·∫°i dataset d·ª±a tr√™n tr·ªçng s·ªë CSD...")
resampled_dataset = random.choices(
    population=dataset,
    weights=final_weights,
    k=len(dataset)
)

# --- 5. Ghi ƒë√® file metadata.jsonl ---
print(f"Ghi ƒë√® file {METADATA_FILE} v·ªõi dataset ƒë√£ ƒë∆∞·ª£c l·∫•y m·∫´u l·∫°i.")
with open(METADATA_FILE, 'w') as outfile:
    for entry in resampled_dataset:
        final_entry = {"file_name": entry["file_name"], "prompt": entry["prompt"]}
        json.dump(final_entry, outfile)
        outfile.write('\n')

# D·ªçn d·∫πp b·ªô nh·ªõ
del text_encoder, tokenizer
gc.collect()
torch.cuda.empty_cache()

print("Ho√†n t·∫•t vi·ªác √°p d·ª•ng CSD!")
```

```python
from diffusers import StableDiffusionXLPipeline
import torch

# --- 1. C·∫•u h√¨nh ---
base_model = "stabilityai/stable-diffusion-xl-base-1.0"

# --- S·ª¨A L·∫†I ƒê∆Ø·ªúNG D·∫™N ·ªû ƒê√ÇY ---
lora_dir = "/kaggle/working/SDXL_LoRA_model"
# T√™n file LoRA m·∫∑c ƒë·ªãnh khi train b·∫±ng script diffusers
lora_name = "pytorch_lora_weights.safetensors"
# --------------------------------

print(f"Loading LoRA from: {lora_dir}/{lora_name}")

# --- 2. Load base model ---
pipe = StableDiffusionXLPipeline.from_pretrained(
    base_model,
    torch_dtype=torch.float16
)
pipe.to("cuda")

# --- 3. Load LoRA ---
# T·∫£i LoRA t·ª´ th∆∞ m·ª•c ƒë√∫ng
pipe.load_lora_weights(
    pretrained_model_name_or_path_or_dict=lora_dir,
    weight_name=lora_name
)

seed = 42 # B·∫°n c√≥ th·ªÉ ch·ªçn b·∫•t k·ª≥ s·ªë n√†o, nh∆∞ng h√£y d√πng chung s·ªë n√†y cho m·ªçi l·∫ßn so s√°nh
generator = torch.Generator("cuda").manual_seed(seed)
# --- 4. Sinh ·∫£nh ---
prompt = "a bathroom with a sink, mirror, and shower. The sink is placed near the center of the room, and the mirror is mounted above it. The shower is visible on the right side of the room, and the shower head is mounted on the wall above the sink.There are multiple towels in the bathroom "
image = pipe(
    prompt,
    num_inference_steps=25,
    guidance_scale=7.0,
    cross_attention_kwargs={"scale": 0.8},
    generator=generator
).images[0]

# --- 5. L∆∞u ·∫£nh ---
image.save("/kaggle/working/output_xl_lora.png")
print("·∫¢nh ƒë√£ ƒë∆∞·ª£c t·∫°o v√† l∆∞u th√†nh c√¥ng t·∫°i /kaggle/working/output_xl_lora.png")

# Hi·ªÉn th·ªã ·∫£nh
display(image)
```
